## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# Exercise 1
import random
import numpy as np

count = 0

# Run simulation 1000 times
for i in range(1000):
    # Roll two dices
    dice_1 = random.randint(1, 6)
    dice_2 = random.randint(1, 6)

    # Sum of the values obtanied
    result = dice_1 + dice_2

    # Count the results that meet the condition
    if (result > 7) or (result % 2 == 0):
        count += 1

# Calculate probability
probability = count / 1000

print(f"The probability of rolling an even number or greater than 7 is: {np.round(probability *100, 2)}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# Initial counters
count_1 = 0
count_2 = 0
n_simmulation = 1000
# Simulation 1000 times
for _ in range(n_simmulation):
    # Take 5 balls with replacement
    balls = [random.choice(list(ball_box.values())) for _ in range(5)]

    # Count result
    if balls.count('White') == 3 and balls.count('Red') == 2:
        count_1 += 1
    if len(set(balls)) == 1:
        count_2 += 1

# Calculate probability
prob_1 = np.round(count_1 / n_simmulation * 100, 2)
prob_2 = np.round(count_2 / n_simmulation * 100, 2)

# Show results

print(f"The probability of taking 3 Whites and 2 reds is: {prob_1}%")
print(f"The probability of taking all balls the same color is: {prob_2}%")
```
